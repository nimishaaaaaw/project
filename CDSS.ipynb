```python
import matplotlib.pyplot as plt
import pandas as pd


df = pd.DataFrame({'Patient Index': range(10), 'CholesterolTotal': [220, 250, 180, 280, 210, 200, 245, 190, 270, 230]})


desirable_threshold = 200
borderline_high_threshold = 240

plt.plot(df['Patient Index'], df['CholesterolTotal'], marker='o', linestyle='-')
plt.axhline(y=desirable_threshold, color='green', linestyle='--', label='Desirable')
plt.axhline(y=borderline_high_threshold, color='orange', linestyle='--', label='Borderline High')
plt.xlabel('Patient Index')
plt.ylabel('Total Cholesterol (mg/dL)')
plt.grid(True)
plt.title('Cholesterol Levels Over Patients')
plt.legend()
plt.show()
```
